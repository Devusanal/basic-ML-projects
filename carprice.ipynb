```python
import pandas as pd
```

```python
car = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/Car%20Price.csv')
```

```python
y = car['Selling_Price']
```

```python
x = car[['Brand', 'Model', 'Year', 'KM_Driven', 'Fuel',
       'Seller_Type', 'Transmission', 'Owner']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
```

```python
# turn the categories into numbers
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


categorical_features = ["Brand" , "Model" , "Fuel", "Seller_Type" , "Transmission" , "Owner"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot" ,
                                  one_hot ,
                                  categorical_features)],
                                  remainder = "passthrough")

transformed_x = transformer.fit_transform(x)
```

```python
pd.DataFrame(transformed_x)
```

```python
dummies = pd.get_dummies(car[["Brand" , "Model" , "Fuel", "Seller_Type" , "Transmission" , "Owner"]])
dummies
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(transformed_x,y,train_size=0.7,random_state=2529)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
